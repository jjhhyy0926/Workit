```python
import json

CONTRACT_PATH = "/content/drive/Othercomputers/내 노트북/project/Workit/data/jihye_sft/contract/augmented/contract_risk_augmented_v4.json"
PLAN_PATH = "/content/drive/Othercomputers/내 노트북/project/Workit/data/jihye_sft/compliance/seed/plan_review_seed_v1.json"
OUT_PATH = "/content/drive/Othercomputers/내 노트북/project/Workit/data/jihye_sft/workit_sft_merged_v2.json"

with open(CONTRACT_PATH, encoding="utf-8") as f:
    contract = json.load(f)

with open(PLAN_PATH, encoding="utf-8") as f:
    plan = json.load(f)

SYSTEM_CONTRACT = "당신은 공공 SW 계약서의 위험 조항을 탐지하는 전문가입니다. 주어진 계약 조항과 참고 기준을 바탕으로 위험 여부를 판단하고 근거를 제시하십시오."
SYSTEM_PLAN = "당신은 공공 SW 사업수행계획서를 검토하는 감리 전문가입니다. 주어진 기준에 따라 항목 내용을 검토하고 충족/미흡/불가 판정과 구체적인 코멘트를 작성하십시오."

def to_exaone_format(item, task):
    system = SYSTEM_CONTRACT if task == "contract" else SYSTEM_PLAN
    return {
        "messages": [
            {"role": "system", "content": system},
            {"role": "user", "content": f"{item['instruction']}\n\n{item['input']}"},
            {"role": "assistant", "content": item["output"]},
        ]
    }

merged = []
for d in contract:
    merged.append(to_exaone_format(d, "contract"))
for d in plan:
    merged.append(to_exaone_format(d, "plan"))

print(f"계약서: {len(contract)}개")
print(f"수행계획서: {len(plan)}개")
print(f"합계: {len(merged)}개")

with open(OUT_PATH, "w", encoding="utf-8") as f:
    json.dump(merged, f, ensure_ascii=False, indent=2)
print(f"저장 완료 → {OUT_PATH}")
```
